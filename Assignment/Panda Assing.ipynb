```python
import pandas as pd
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
wk_path = os.getcwd()

allFiles = glob.glob(wk_path + "/*.csv")
us_census = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_)
    list_.append(df)
us_census = pd.concat(list_,axis=0).reset_index()
us_census
```

```python
del(us_census['index']),
del(us_census['Unnamed: 0']),
```

```python
us_census['Hispanic']= us_census['Hispanic'].replace('%','',regex=True).astype('float').drop_duplicates()
us_census['Hispanic']= us_census['Hispanic'].fillna(0)
us_census['Black']= us_census['Black'].replace('%','',regex=True).astype('float').drop_duplicates()
us_census['Black']= us_census['Black'].fillna(0)
us_census['Native']= us_census['Native'].replace('%','',regex=True).astype('float').drop_duplicates()
us_census['Native']= us_census['Native'].fillna(0)
us_census['Asian']= us_census['Asian'].replace('%','',regex=True).astype('float').drop_duplicates()
us_census['Asian']= us_census['Asian'].fillna(0)
us_census['Pacific']= us_census['Pacific'].replace('%','',regex=True).astype('float').drop_duplicates()
us_census['Pacific']= us_census['Pacific'].fillna(0)
us_census['Income']= us_census['Income'].astype('float').drop_duplicates()
us_census['Income']= us_census['Income'].fillna(0)
us_census['White']= us_census['White'].replace('%','',regex=True).astype('float').drop_duplicates()
us_census['White']= us_census['White'].fillna(0)
us_census['TotalPop']= us_census['TotalPop'].astype('float').drop_duplicates()
us_census['TotalPop']= us_census['TotalPop'].fillna(0)

```
